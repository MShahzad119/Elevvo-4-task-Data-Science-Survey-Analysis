# Data Cleaning

```python
df.drop_duplicates(inplace=True)
```

```python
# Fill numeric NaN with median
for col in df.select_dtypes(include=['int64', 'float64']):
    df[col] = df[col].fillna(df[col].median())
```

```python
# Fill categorical NaN with 'Unknown'
for col in df.select_dtypes(include=['object']):
    df[col] = df[col].fillna("Unknown")
```

```python
df['Q3'] = df['Q3'].str.strip().str.title()
```

# Handle Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

df = df.copy()  # defragment memory
encoder = LabelEncoder()
df['Q3_encoded'] = encoder.fit_transform(df['Q3'])
```

```python
country_map = {'United States': 1, 'India': 2, 'Other': 3}
df['Q3_mapped'] = df['Q3'].map(country_map)
```

# Extract Insights

## Top 5 programming languages:

```python
# Top 5 programming languages
top_languages = pd.Series(df.filter(like='Q7_Part').stack().dropna()).value_counts().head(5)
print(top_languages)
```

## Average coding experience by country:

```python
# Know experience mapping
df['Q6_num'] = df['Q6'].map({
    '< 1 years': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    '5-10 years': 7.5,
    '10-20 years': 15,
    '20+ years': 25
})

# Average coding experience by mapped country
avg_exp_country = df.groupby('Q3_mapped')['Q6_num'].mean()
print(avg_exp_country)
```

# Create a Dashboard or Chart (Bonus)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Top 5 programming languages
plt.figure(figsize=(8,5))
sns.barplot(
    x=top_languages.values,
    y=top_languages.index,
    hue=top_languages.index,  # hue specify kiya
    palette="viridis",
    legend=False              # legend hide kiya
)
plt.xlabel("Number of Respondents")
plt.ylabel("Programming Language")
plt.title("Top 5 Programming Languages")
plt.show()
```

```python
plt.figure(figsize=(8,5))
country_counts = df['Q3'].value_counts().head(5)

sns.barplot(
    x=country_counts.values,
    y=country_counts.index,
    hue=country_counts.index,
    palette="coolwarm",
    legend=False
)
plt.xlabel("Number of Respondents")
plt.ylabel("Country")
plt.title("Top 5 Countries by Respondents")
plt.show()
```

```python
plt.figure(figsize=(10,5))
edu_counts = df['Q4'].value_counts()

sns.barplot(
    x=edu_counts.values,
    y=edu_counts.index,
    hue=edu_counts.index,
    palette="magma",
    legend=False
)
plt.xlabel("Number of Respondents")
plt.ylabel("Education Level")
plt.title("Education Level Distribution")
plt.show()
```
